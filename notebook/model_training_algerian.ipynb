```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv("Algerian_cleaned.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
##drop month,day,year
df.drop(['day','month','year'],axis=1,inplace=True)
df.head()
```

```python
###we convert the vategorical variable(classes) into numerical variable by inntitialseing 0 and 1
```

```python
df['Classes'].value_counts()
```

```python
df['Classes']=np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df.tail()
```

```python
df['Classes'].value_counts()
```

```python
##we are tying to calculate FWI-fire wheather index 
##the out feature is FWI
X=df.drop('FWI',axis=1)
y=df['FWI']
```

```python
y
```

```python
##train test  split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape
```

```python
X_train.corr()
```

```python
##check for multicolinearity
plt.figure(figsize=(12,10))
corr=X_train.corr()
sns.heatmap(corr,annot=True)
```

```python
##the independent 

def correlation(dataset,threshold):
    col_corr=set() 
    corr_matrix=dataset.corr()
    for  i in range(len(corr_matrix.columns)):##through each columns
        for j in range(i): ##iterating each element in the column
            if abs(corr_matrix.iloc[i,j])>threshold:
                ##abs will make any -ve corr to +ve
                colname=corr_matrix.columns[i]
                ##if threshold is greater we are gonna take the column and save it


                ##this i,j is the index of the column
                #example temerature[i] and temperature[j]

                ##probably  try tocheck wtr all columns

                col_corr.add(colname)

    return col_corr
```

```python
corr_features=correlation(X_train,0.85)
```

```python
corr_features
##we can drop th correlation features
```

```python
X_train.shape
```

```python
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
X_train.shape,X_test.shape
```

##feature scaling or  standardization

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
X_train_scaled=scalar.fit_transform(X_train)
X_test_scaled=scalar.transform(X_test)
```

```python
X_train_scaled
```

BOX PLOTS TO UNDERSTAND EFFECT OF STANDARD SCALAR


```python
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title('X_train before scaling')
plt.subplot(1,2,2)
sns.boxplot(data=X_test_scaled)
plt.title('X_train after scaling')

##this is y scaling is imp
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
linreg=LinearRegression()
linreg.fit(X_train_scaled,y_train)
y_pred=linreg.predict(X_test_scaled)
mae=mean_absolute_error(y_test)
```

```python
score=r2_score(y_test,y_pred)
plt.scatter(y_test,y_pred)
```

```python
##very good model since it is linear model

```

```python
##lasso regression default alpha =0.25
from sklearn.linear_model import Lasso
lasso=Lasso()
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

lasso.fit(X_train_scaled,y_train)
y_pred=lasso.predict(X_test_scaled)
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
plt.scatter(y_test,y_pred)

print("mse:",mse)
print("mae:",mae)
print("r2 score:",score)
```

```python
##lasso cross validation
from sklearn.linear_model import LassoCV
lassocv=LassoCV(cv=5)
lassocv.fit(X_train_scaled,y_train)
```
